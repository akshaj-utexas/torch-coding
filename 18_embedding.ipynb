# 🟢 Easy: Embedding Layer

Implement an **embedding lookup table** from scratch.

### Signature
```python
class MyEmbedding(nn.Module):
    def __init__(self, num_embeddings: int, embedding_dim: int): ...
    def forward(self, indices: Tensor) -> Tensor: ...
```

### Rules
- `self.weight`: `nn.Parameter` of shape `(num_embeddings, embedding_dim)`
- Forward: index into weight matrix — `weight[indices]`
- Do NOT use `nn.Embedding`

````python
import torch
import torch.nn as nn
````

````python
# ✏️ YOUR IMPLEMENTATION HERE

class MyEmbedding(nn.Module):
    def __init__(self, num_embeddings, embedding_dim):
        super().__init__()
        # save the arguments
        self.num_embeddings = num_embeddings
        self.embedding_dim = embedding_dim
        # we want to create a weight matrix with nn.Parameter, but nn.Parameter requires tensors
        # nn.Parameter wraps itself around those tensors
        # so we require tensors of shape (num_embeddings, embedding_dim).
        # torch.randn(num_embeddings, embedding_dim) will make the tensors with numbers that are normally distributed (mean 0 var 1)
        # torch.rand_like(x) cannot be used since it is used to make random number tensors that has the same shape as the original tensor x
        self.weight = nn.Parameter(torch.randn(num_embeddings, embedding_dim))

    def forward(self, indices):
        # we are essentially just looking up weights given some indices.
        return self.weight[indices]
````

````python
# 🧪 Debug
emb = MyEmbedding(10, 4)
idx = torch.tensor([0, 3, 7])
print('Output shape:', emb(idx).shape)
print('Matches manual:', torch.equal(emb(idx)[0], emb.weight[0]))
````
