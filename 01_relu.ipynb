# 🟢 Easy: Implement ReLU

Implement the **ReLU** (Rectified Linear Unit) activation function from scratch.

$$\text{ReLU}(x) = \max(0, x)$$

### Signature
```python
def relu(x: torch.Tensor) -> torch.Tensor:
    ...
```

### Rules
- Do **NOT** use `torch.relu`, `F.relu`, `torch.clamp`, or any built-in activation
- Must support autograd (gradients should flow back)

### Example
```
Input:  tensor([-2., -1., 0., 1., 2.])
Output: tensor([ 0.,  0., 0., 1., 2.])
```

````python
# ✏️ YOUR IMPLEMENTATION HERE

def relu(x: torch.Tensor) -> torch.Tensor:
  # create a mask that checks for all values in x if they are greater than 0 (True if yes False otherwise)
  mask = (x > 0) # this makes a tensor mask that has [True, False] for all entires of x

  # we then multiply this mask with the tensor, False is treated as 0, True is treated as 1.
  # so negative values are zeroed, while positive are left as they are

  # e.g. x = [-1, 0, 5, -2, 4]
  # mask = (x > 0): mask = [False, False, True, False, True]
  # False = 0; True = 1
  # ReLU = x * mask = [-1 * False, 0 * False, 5 * True, -2 * False, 4 * True] = [0, 0, 5, 0, 4]

  return x * mask
    # pass  # Replace this
````
