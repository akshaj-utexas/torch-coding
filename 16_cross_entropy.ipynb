# 🟢 Easy: Cross-Entropy Loss

Implement **cross-entropy loss** from scratch.

$$\text{CE}(x, y) = -\log\frac{e^{x_y}}{\sum_j e^{x_j}}$$

### Signature
```python
def cross_entropy_loss(logits: Tensor, targets: Tensor) -> Tensor:
    # logits: (B, C) float, targets: (B,) long indices
    # Returns: scalar loss (mean over batch)
```

### Rules
- Do NOT use `F.cross_entropy` or `nn.CrossEntropyLoss`
- Must be numerically stable (use logsumexp trick)

````python
import torch
````

````python
# ✏️ YOUR IMPLEMENTATION HERE

def cross_entropy_loss(logits, targets):
  # the exponential function e^x grows very fast, e^1000 = inf
  # if logits are a little big or small, e^x will hit infinity/go to zero, breaking loss calculation.
  # logsumexp removes a constant a from every input, and adds it back at the end.
  # we choose max(x) to subtract from each term, then add it back.

  # results in no overflow, and no underflow (atleast one term is 0, therefore e^0 = 1)
  lse = torch.logsumexp(logits, dim=-1)
  rows = torch.arange(logits.size(0))

  correct_logits = logits[rows, targets]

  return torch.mean(lse - correct_logits)
````

````python
# 🧪 Debug
logits = torch.randn(4, 10)
targets = torch.randint(0, 10, (4,))
print('Loss:', cross_entropy_loss(logits, targets))
print('Ref: ', torch.nn.functional.cross_entropy(logits, targets))
````
