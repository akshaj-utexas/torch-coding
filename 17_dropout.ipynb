# 🟢 Easy: Implement Dropout

Implement **Dropout** regularization from scratch.

### Signature
```python
class MyDropout(nn.Module):
    def __init__(self, p: float = 0.5): ...
    def forward(self, x: Tensor) -> Tensor: ...
```

### Rules
- During **training**: zero each element with probability `p`, scale remaining by `1/(1-p)`
- During **eval**: return input unchanged (identity)
- Do NOT use `nn.Dropout` or `F.dropout`

````python
import torch
import torch.nn as nn
````

````python
# ✏️ YOUR IMPLEMENTATION HERE

class MyDropout(nn.Module):
    def __init__(self, p=0.5):
        super().__init__()
        self.p = p

    def forward(self, x):
        if not self.training or self.p == 0:
          return x

        # Create a binary mask
        # torch.rand_like(x) generates random numbers between 0 and 1
        # (mask > self.p) gives us 1s where we keep and 0s where we drop
        # makes trues and false, trues where rand_like has a value of more than p,
        # so we essentially keep all the random numbers that are more than p, hence keeping a subset of the numbers
        mask = torch.rand_like(x) > self.p
        mask.float() # true - 1, false - 0

        return (x * mask) / (1-self.p) # scale by 1-p

````

````python
# 🧪 Debug
d = MyDropout(p=0.5)
d.train()
x = torch.ones(10)
print('Train:', d(x))
d.eval()
print('Eval: ', d(x))
````
