# 🟢 Easy: Implement Dropout

Implement **Dropout** regularization from scratch.

### Signature
```python
class MyDropout(nn.Module):
    def __init__(self, p: float = 0.5): ...
    def forward(self, x: Tensor) -> Tensor: ...
```

### Rules
- During **training**: zero each element with probability `p`, scale remaining by `1/(1-p)`
- During **eval**: return input unchanged (identity)
- Do NOT use `nn.Dropout` or `F.dropout`

````python
import torch
import torch.nn as nn
````

````python
# ✏️ YOUR IMPLEMENTATION HERE

class MyDropout(nn.Module):
    def __init__(self, p=0.5):
        super().__init__()
        # store propability
        self.p = p

    def forward(self, x):
        # self.training is a pre-built flag to know if we are in training or eval.
        # if training, we need to randomly set weights to 0
        # same as ReLU, just that we set whatever is less than p to 0 (False). Anything more than p is set to 1 (True).
        # in ReLU, we compared with 0, here we compare with p
        # multiply x with this mask
        # also scale remaining weights so that they don't become too dominant
        if self.training:
            mask = torch.rand_like(x)
            # if random number greater than p, keep. If small, set to 0
            keep = mask > self.p
            return (x * keep) / (1 - self.p)
        else:
            return x
````

````python
# 🧪 Debug
d = MyDropout(p=0.5)
d.train()
x = torch.ones(10)
print('Train:', d(x))
d.eval()
print('Eval: ', d(x))
````
