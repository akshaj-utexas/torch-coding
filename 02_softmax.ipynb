# 🟢 Easy: Implement Softmax

Implement the **Softmax** function from scratch.

$$\text{softmax}(x_i) = \frac{e^{x_i}}{\sum_j e^{x_j}}$$

### Signature
```python
def my_softmax(x: torch.Tensor, dim: int = -1) -> torch.Tensor:
    ...
```

### Rules
- Do **NOT** use `torch.softmax`, `F.softmax`, or `torch.nn.Softmax`
- Must be **numerically stable** (hint: subtract `max` before `exp`)

### Example
```
Input:  tensor([1., 2., 3.])
Output: tensor([0.0900, 0.2447, 0.6652])  # sums to 1.0
```

````python
# ✏️ YOUR IMPLEMENTATION HERE

def my_softmax(x: torch.Tensor, dim: int = -1) -> torch.Tensor:

  # we want to find highest value among all entries, then we want to only pick the values
  # torch.max returns two things at once, .values (actual max num),
  # and .indices (where it was found), we only care about the number
  # keep_dim ensures that we broadcast to all rows, so if a 2d matrix [5, 10] is given,
  # keep_dim=False will return [1], but keep_dim=True will return [5,1] so we can easily subtract later on

  max_x = torch.max(x, dim=dim, keepdim=True).values
  print(f"max_x shape: {max_x.shape}")
  numerator = torch.exp(x - max_x)
  denominator = torch.sum(torch.exp(x - max_x), dim=dim, keepdim=True)
  return numerator/denominator
````

````python
# 🧪 Debug
x = torch.tensor([1.0, 2.0, 3.0])
print("Output:", my_softmax(x, dim=-1))
print("Sum:   ", my_softmax(x, dim=-1).sum())  # should be ~1.0
print("Ref:   ", torch.softmax(x, dim=-1))
````
